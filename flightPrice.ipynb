#### **Feature Engineering**

##### Date_of_Journey

```python
df['Date'] = df['Date_of_Journey'].str.split('/').str[0].astype(int)
df['Month'] = df['Date_of_Journey'].str.split('/').str[1].astype(int)
df['Year'] = df['Date_of_Journey'].str.split('/').str[2].astype(int)
df.drop('Date_of_Journey', axis=1, inplace=True)
```

```python
df.head()
```

```python
df.info()
```

##### Arrival_Time

```python
arrival_time_only = df['Arrival_Time'].str.split(' ').str[0]
df['Arrival_hour'] = arrival_time_only.str.split(':').str[0].astype(int)
df['Arrival_minute'] = arrival_time_only.str.split(':').str[1].astype(int)
df.drop('Arrival_Time', axis=1, inplace=True)
```

```python
df.head()
```

```python
df.info()
```

##### Dep_Time

```python
df['Departure_hour'] = df['Dep_Time'].str.split(':').str[0].astype(int)
df['Departure_minute'] = df['Dep_Time'].str.split(':').str[1].astype(int)
df.drop('Dep_Time', axis=1, inplace=True)
```

```python
df.head()
```

##### Duration

```python
df['Duration'] = df['Duration'].apply(lambda x: '0h 0m' if len(x.split()) == 1 else x)
df['Duration_hour'] = df['Duration'].str.split(' ').str[0].str.replace('h', "").astype(int)
df['Duration_minute'] = df['Duration'].str.split(' ').str[1].str.replace('m', "").astype(int)
df.drop('Duration', axis=1, inplace=True)
```

```python
df['Total_Stops'] = df['Total_Stops'].str.strip().str.lower()

df['Total_Stops'] = df['Total_Stops'].map({
    'non-stop': 0,
    '1 stop': 1,
    '2 stops': 2,
    '3 stops': 3,
    '4 stops': 4,
}).fillna(1)
```
